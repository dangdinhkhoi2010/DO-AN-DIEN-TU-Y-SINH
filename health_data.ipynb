```python
#1. TẠO KẾT NỐI VỚI CỔNG COM4
import serial                                             # Thêm thư viện serial để Python đọc dữ liệu từ ESP8266 gửi qua COM4
import csv                                                # Thêm thư viện csv để ghi dữ liệu vào file .csv
import time                                               # Thêm thư viện time để xử lý thời gian
ser = serial.Serial('COM4', 115200)                       # Mở kết nối serial tại COM4 với tốc độ baudrate 115200
time.sleep(2)                                             # Dừng 2 giây để ESP8266 ổn định, có thời gian reset và chuẩn bị gửi dữ liệu

#2. TẠO FILE health_data.csv VÀ GHI TIÊU ĐỀ
filename = "health_data.csv"                              # Đặt tên file dữ liệu là health_data.csv
with open(filename, mode='w', newline='', buffering=1) as file: 
# Mở file health_data.csv ở chế độ ghi ('w'), đặt newline='' để mỗi dòng chỉ có 1 dòng mới và buffering=1 để dữ liệu được ghi ngay tại từng dòng
   
    writer = csv.writer(file)                             # Tạo đối tượng writer để ghi dữ liệu vào file csv theo dạng bảng (hàng, cột)
    writer.writerow(["HeartRate", "SpO2"])                # Ghi dòng đầu tiên có tiêu đề HeartRate và SpO2 vào file health_data.csv
    print("Đang ghi dữ liệu... Nhấn Ctrl+C để dừng ghi!") # In ra thông báo đang ghi dữ liệu, muốn dừng ghi thì ấn Ctrl+C

#3. ĐỌC DỮ LIỆU LIÊN TỤC TỪ CỔNG COM4, GIẢI MÃ VÀ KIỂM TRA
    try:
        while True:                                       # Tạo 1 vòng lặp vô hạn để đọc dữ liệu liên tục
            raw = ser.readline()                          # Đọc 1 dòng dữ liệu từ cổng COM4
            try:
                line = raw.decode('utf-8').strip()        # Chuyển raw (dữ liệu dạng byte) thành string (chuỗi kí tự)
                                                          # .strip() để xoá các kí tự thừa (\n, \r, khoảng trắng...)
                print(f"Dữ liệu nhận được: {line}")       # In ra dòng dữ liệu vừa nhận được từ ESP8266
                data = line.split(',')                    # Cắt chuỗi line thành một danh sách gồm hai phần, tách theo dấu phẩy
                                                          # Ví dụ: "78,97" => ['78', '97']
                if len(data) == 2:                        # Kiểm tra xem có đúng 2 phần tử nhịp tim và SpO2 không, nếu sai thì bỏ qua (không ghi)
                    heart_rate = float(data[0])           # Chuyển chuỗi đầu tiên (nhịp tim) sang số thực
                    spo2 = float(data[1])                 # Chuyển chuỗi thứ hai (SpO2) sang số thực
                    writer.writerow([heart_rate, spo2])   # Ghi kết quả vào một hàng mới trong file health_data.csv                    
            except UnicodeDecodeError:
                print("Bỏ qua dòng lỗi mã hóa!")          # Nếu có dòng nào bị lỗi giải mã (dữ liệu không đúng UTF-8) thì bỏ qua và in thông báo
    except KeyboardInterrupt:
        print("Kết thúc ghi dữ liệu!")                    # Khi nhấn Ctrl+C thì thoát khỏi vòng lặp và thông báo kết thúc ghi dữ liệu
```
